```python
from pycocotools.coco import COCO
import glob
import json
from torchvision.ops import box_iou
import torch
from detectron2.structures import BoxMode
from tqdm.auto import tqdm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from collections import defaultdict
from pprint import pprint
from scipy.signal import medfilt
import os
import copy
from PIL import Image
```

```python
gt_path = "../data/val_annot_pos_neg_strict_frame_cocofied_vc.json"
gt_coco = COCO(gt_path)
```

```python
dt_paths = sorted(glob.glob("/scratch/shared/beegfs/prannay/ego4d_data/ckpt/perframe_run/inference_vq2d_val_pos_neg_strict_frame_vc/*.json"))
```

```python
detections = []
for dt_path in dt_paths:
    with open(dt_path, "r") as fp:
        detections += json.load(fp)
```

```python
dt_coco = gt_coco.loadRes(detections)
```

```python
pos_iids = set(gt_coco.getImgIds(catIds=[1]))
```

```python
# lets get the frame numbers for each dataset_uid
duid2fnos = defaultdict(list)
for iid in gt_coco.imgs:
    duid = gt_coco.imgs[iid]["dataset_uid"]
    fno = int(os.path.basename(gt_coco.imgs[iid]["file_name"]).split("_")[1].split(".")[0])
    duid2fnos[duid].append(fno)
```

```python
duids = set([gt_coco.imgs[iid]["dataset_uid"] for iid in gt_coco.imgs])
duid2iids = defaultdict(set)
for iid in gt_coco.imgs:
    duid = gt_coco.imgs[iid]["dataset_uid"]
    duid2iids[duid].add(iid)
```

```python
# get iids for each duid in frame order
duid2iids_ordered = {}
for duid in duids:
    iids = duid2iids[duid]
    imgs = gt_coco.loadImgs(iids)
    iids = [img['id'] for img in sorted(imgs, key=lambda x: int(os.path.basename(x["file_name"]).split("_")[1].split(".")[0]))]
    duid2iids_ordered[duid] = iids
```

```python
pprint([dt_coco.loadImgs(a)[0]['file_name'] for a in duid2iids_ordered[duid]])
```

```python
# get the max score and ann_id for each frame from duid2iids_ordered
duid2highest = defaultdict(list)
for duid, iids in tqdm(duid2iids_ordered.items(), total=len(duid2iids_ordered)):
    # print(iids)
    for iid in iids:
        dt_ann_ids = dt_coco.getAnnIds(imgIds=iid)
        dt_anns = dt_coco.loadAnns(dt_ann_ids)
        assert len(dt_anns)
        gt_ann_ids = gt_coco.getAnnIds(imgIds=iid)
        pos_frame = len(gt_ann_ids) > 0
        dt_anns = sorted(dt_anns, key=lambda x: x['score'], reverse=True)
        top_ann = copy.deepcopy(dt_anns[0])
        top_ann['positive'] = pos_frame
        # check iou with gt if pos_frame
        if pos_frame:
            gt_ann = gt_coco.loadAnns(gt_ann_ids)[0]
            bbox_gt = gt_ann['bbox']
            bbox_gt = BoxMode.convert(bbox_gt, BoxMode.XYWH_ABS, BoxMode.XYXY_ABS)
            bbox_dt = top_ann['bbox']
            bbox_dt = BoxMode.convert(bbox_dt, BoxMode.XYWH_ABS, BoxMode.XYXY_ABS)
            iou = box_iou(torch.tensor([bbox_gt]), torch.tensor([bbox_dt]))[0][0].item()
            top_ann['iou'] = iou
        else:
            top_ann['iou'] = -1.0
        duid2highest[duid].append(top_ann)
```

```python
def plot_score_trace(duid2highest, gt_coco, duid, iou_thresh=0.5):
    file_names = [gt_coco.loadImgs(a)[0]['file_name'] for a in duid2iids_ordered[duid]]
    fnos = [int(os.path.basename(a).split("_")[1].split(".")[0]) for a in file_names]
    scores = [a['score'] for a in duid2highest[duid]]
    scores_med = medfilt(scores, kernel_size=5).tolist()
    # print(scores)
    pos = [a['positive'] for a in duid2highest[duid]]
    correct = [a['iou'] > iou_thresh for a in duid2highest[duid]]
    scores_true_pos = [sc for sc, p, c in zip(scores, pos, correct) if p and c]
    scores_missed_pos = [sc for sc, p, c in zip(scores, pos, correct) if p and not c]
    scores_negative = [sc for sc, p, c in zip(scores, pos, correct) if not p]

    scores_med_true_pos = [sc for sc, p, c in zip(scores_med, pos, correct) if p and c]
    scores_med_missed_pos = [sc for sc, p, c in zip(scores_med, pos, correct) if p and not c]
    scores_med_negative = [sc for sc, p, c in zip(scores_med, pos, correct) if not p]

    fnos_true_pos = [fno for fno, p, c in zip(fnos, pos, correct) if p and c]
    fnos_missed_pos = [fno for fno, p, c in zip(fnos, pos, correct) if p and not c]
    fnos_negative = [fno for fno, p, c in zip(fnos, pos, correct) if not p]
    fig, ax = plt.subplots(2, 1, figsize=(20, 20))
    ax[0].plot(fnos, scores, c="k")
    ax[0].plot(fnos_true_pos, scores_true_pos, c="g", marker="o", ls="None")
    ax[0].plot(fnos_missed_pos, scores_missed_pos, c="r", marker="o", ls="None")
    ax[0].plot(fnos_negative, scores_negative, c="r", marker="x", ls="None")

    ax[1].plot(fnos, scores_med, c="k")
    ax[1].plot(fnos_true_pos, scores_med_true_pos, c="g", marker="o", ls="None")
    ax[1].plot(fnos_missed_pos, scores_med_missed_pos, c="r", marker="o", ls="None")
    ax[1].plot(fnos_negative, scores_med_negative, c="r", marker="x", ls="None")

    # save the plot
    # med_tog = "med" if med_filt else "raw"
    file_name = f"score_trace_{duid}.pdf"
    directory = "/scratch/shared/beegfs/prannay/ego4d_data/ckpt/perframe_run/inference_vq2d_val_pos_neg_strict_frame_vc/traces/"
    plt.tight_layout()
    plt.savefig(os.path.join(directory, file_name))
    plt.close()
```

```python
import warnings
warnings.filterwarnings("ignore")
for duid in tqdm(sorted(duids), total=len(duids)):
    plot_score_trace(duid2highest, gt_coco, duid, iou_thresh=0.5)
warnings.filterwarnings("default")
```

```python
iid2iou_frame = {}
for iid in tqdm(pos_iids, total=len(pos_iids)):
    bbox_gt = torch.tensor([ann["bbox"] for ann in gt_coco.imgToAnns[iid]])
    det_scores = [ann["score"] for ann in dt_coco.imgToAnns[iid]]
    aids = [ann['id'] for ann in dt_coco.imgToAnns[iid]]
    bbox_dt = torch.tensor(
        [x["bbox"] for x, _ in sorted(zip(dt_coco.imgToAnns[iid], det_scores), key=lambda x: x[1], reverse=True)]
    )
    aids = [x for _, x in sorted(zip(det_scores, aids), key=lambda x: x[0], reverse=True)]
    bbox_gt = BoxMode.convert(bbox_gt, BoxMode.XYWH_ABS, BoxMode.XYXY_ABS)
    bbox_dt = BoxMode.convert(bbox_dt, BoxMode.XYWH_ABS, BoxMode.XYXY_ABS)
    ious = box_iou(bbox_gt, bbox_dt)
    assert ious.size(0) == 1
    iid2iou_frame[iid] = list(zip(ious[0].tolist(), aids))
```

```python
duid2iid = defaultdict(set)
duid2iou_video = defaultdict(list)
pos_imgs = gt_coco.loadImgs(pos_iids)
for img in pos_imgs:
    duid2iid[img["dataset_uid"]].add(img["id"])

for duid, iids in tqdm(duid2iid.items(), total=len(duid2iid)):
    for iid in iids:
        bbox_gt = torch.tensor([ann["bbox"] for ann in gt_coco.imgToAnns[iid]])
        det_scores = [ann["score"] for ann in dt_coco.imgToAnns[iid]]
        aids = [ann['id'] for ann in dt_coco.imgToAnns[iid]]
        bbox_dt = torch.tensor([x["bbox"] for x in dt_coco.imgToAnns[iid]])
        bbox_gt = BoxMode.convert(bbox_gt, BoxMode.XYWH_ABS, BoxMode.XYXY_ABS)
        bbox_dt = BoxMode.convert(bbox_dt, BoxMode.XYWH_ABS, BoxMode.XYXY_ABS)
        ious = box_iou(bbox_gt, bbox_dt)
        assert ious.size(0) == 1
        duid2iou_video[duid] += list(zip(ious[0].tolist(), aids, det_scores))

for duid in sorted(duid2iou_video.keys()):
    duid2iou_video[duid] = [(x[0], x[1]) for x in sorted(duid2iou_video[duid], key=lambda x: x[-1], reverse=True)]
```

```python
duid2iou_video_all = defaultdict(list)
duid2iid_all = defaultdict(set)
imgs = gt_coco.loadImgs(gt_coco.getImgIds())
for img in imgs:
    duid2iid_all[img["dataset_uid"]].add(img["id"])

for duid, iids in tqdm(duid2iid_all.items(), total=len(duid2iid_all)):
    for iid in iids:
        bbox_gt = torch.tensor([ann["bbox"] for ann in gt_coco.imgToAnns[iid]])
        det_scores = [ann["score"] for ann in dt_coco.imgToAnns[iid]]
        aids = [ann['id'] for ann in dt_coco.imgToAnns[iid]]
        bbox_dt = torch.tensor([x["bbox"] for x in dt_coco.imgToAnns[iid]])
        if bbox_gt.size(0) == 0:
            # do stuff
            duid2iou_video_all[duid] += list(zip([0.0] * bbox_dt.size(0), det_scores))
        else:
            bbox_gt = BoxMode.convert(bbox_gt, BoxMode.XYWH_ABS, BoxMode.XYXY_ABS)
            bbox_dt = BoxMode.convert(bbox_dt, BoxMode.XYWH_ABS, BoxMode.XYXY_ABS)
            ious = box_iou(bbox_gt, bbox_dt)
            assert ious.size(0) == 1
            duid2iou_video_all[duid] += list(zip(ious[0].tolist(), aids, det_scores))

for duid in sorted(duid2iou_video_all.keys()):
    duid2iou_video_all[duid] = [(x[0], x[1]) for x in sorted(duid2iou_video_all[duid], key=lambda x: x[-1], reverse=True)]
```

```python
# calculate the recall at various K
def get_recall_at_k(iou_dict, k=1, iou=0.5, pbar=True):
    recalls = []
    # if pbar true then use tqdm
    if pbar:
        for ious in tqdm(iou_dict.values(), total=len(iou_dict)):
            recalls.append(np.max(ious[:k]) >= iou)
    else:
        for ious in iou_dict.values():
            recalls.append(np.max(ious[:k]) >= iou)
    return np.mean(recalls)
```

```python
iou_thresholds = [0.3, 0.5, 0.75]
ks = [1, 2, 5, 10, 20, 50, 100, 128]
recalls_frame = np.zeros((len(iou_thresholds), len(ks)))
recalls_video = np.zeros((len(iou_thresholds), len(ks)))
recalls_video_all = np.zeros((len(iou_thresholds), len(ks)))
for i, iou in enumerate(iou_thresholds):
    for j, k in enumerate(ks):
        recalls_frame[i, j] = get_recall_at_k(iid2iou_frame, k=k, iou=iou, pbar=False)
        recalls_video[i, j] = get_recall_at_k(duid2iou_video, k=k, iou=iou, pbar=False)
        recalls_video_all[i, j] = get_recall_at_k(duid2iou_video_all, k=k, iou=iou, pbar=False)
```

```python
# lets plot the recall vs k for different iou thresholds using log scale for the k axis
# use different line styles and marker styles for different iou thresholds
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
linestyles = ['-', '--', ':', '-.']
for i, iou in enumerate(iou_thresholds):
    ax.plot(ks, recalls_video_all[i], label=f"iou={iou}", linestyle=linestyles[i], marker="o", lw=3, markersize=10)
# make text a bit larger
ax.tick_params(axis='both', which='major', labelsize=15)
ax.set_xscale("log")
ax.set_xlabel("k")
ax.set_ylabel("recall")
# Make legend text a bit larger
ax.legend(prop={'size': 15})
ax.set_ylim(0, 1.1)
plt.show()
```

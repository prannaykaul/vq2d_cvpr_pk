```python
duid2pos_frames = defaultdict(set)
for duid, gt in tqdm(duid2gt.items(), total=len(duid2gt)):
    if gt['clip_uid'] is None:
        continue
    for frame in gt['response_track']:
        duid2pos_frames[duid].add(frame['frame_number'])
```

```python
duid2available_frames = defaultdict(set)
for duid, gt in tqdm(duid2gt.items(), total=len(duid2gt)):
    clipid = gt['clip_uid']
    if clipid is None:
        continue
    path_to_frames = os.path.join("/scratch/shared/beegfs/prannay/ego4d_data/images_val/{clipid}/*.png".format(clipid=clipid))
    available_frames = sorted(glob.glob(path_to_frames))
    frame_nos = [int(os.path.basename(f).split("_")[-1].split(".")[0]) - 1 for f in available_frames]
    duid2available_frames[duid].update(frame_nos)
```

```python
print(clipid, frame_nos)
print(gt.keys())
```

```python
duid2neg_frames_strict = defaultdict(set)
duid2neg_frames_loose = defaultdict(set)
for duid, gt in tqdm(duid2gt.items(), total=len(duid2gt)):
    if gt['clip_uid'] is None:
        continue
    pos_frames = duid2pos_frames[duid]
    available_frames = duid2available_frames[duid]
    loose_neg_frames = [f for f in iter(available_frames) if (f not in pos_frames)]
    duid2neg_frames_loose[duid].update(loose_neg_frames)
    strict_neg_frames = [f for f in loose_neg_frames if (f < gt['query_frame']) and (f > max(pos_frames))]
    duid2neg_frames_strict[duid].update(strict_neg_frames)
```

```python
# stats for num of pos and neg frames
num_pos_frames = [len(v) for v in duid2pos_frames.values()]
num_neg_frames_strict = [len(v) for v in duid2neg_frames_strict.values()]
num_neg_frames_loose = [len(v) for v in duid2neg_frames_loose.values()]
print("pos_frames", np.mean(num_pos_frames), np.std(num_pos_frames), np.sum(num_pos_frames))
print("neg_strict", np.mean(num_neg_frames_strict), np.std(num_neg_frames_strict), np.sum(num_neg_frames_strict))
print("neg_loose", np.mean(num_neg_frames_loose), np.std(num_neg_frames_loose), np.sum(num_neg_frames_loose))
```

Plan is to have a COCO file with an additional key in the dict which is a list of visual crops with file names bbox and dataset uid
then in the new dataloader we will need to load the reference file also, hopefully can just add that to dataset mapper

```python
#lets create coco-style annotations from gt_data and save it
coco_gt_data_annotations = []
coco_gt_data_images = []
coco_gt_data_visual_crops = []
aid_counter = 1
iid_counter = 1
vcid_counter = 1
for i, a in tqdm(enumerate(gt_data), total=len(gt_data)):
    # print(a)
    # print(a.keys())
    clip_uid = a['clip_uid']
    if clip_uid is None:
        continue
    coco_gt_data_visual_crops.append({
        'id': vcid_counter,
        'dataset_uid': a['dataset_uid'],
        'bbox': [a['visual_crop']['x'], a['visual_crop']['y'], a['visual_crop']['width'], a['visual_crop']['height']],
        'width': a['visual_crop']['original_width'],
        'height': a['visual_crop']['original_height'],
        'file_name': os.path.join(a['clip_uid'], "frame_{:07d}.png".format(a['visual_crop']['frame_number']+1)),
    })
    if a['clip_uid'] is None:
        continue
    for bb in a['response_track']:
        coco_gt_data_annotations.append({
            'id': aid_counter,
            'image_id': iid_counter,
            'visual_crop_id': vcid_counter,
            'category_id': 1,
            'bbox': [bb['x'], bb['y'], bb['width'], bb['height']],
            'area': bb['width'] * bb['height'],
            'iscrowd': 0,
        })
        path = os.path.join(a['clip_uid'], "frame_{:07d}.png".format(bb['frame_number']+1))
        assert os.path.isfile(os.path.join("/work/prannay/ego4d/images_val", path)), path
        coco_gt_data_images.append({
            'id': iid_counter,
            'visual_crop_id': [vcid_counter],
            'width': bb['original_width'],
            'height': bb['original_height'],
            'file_name': path,
            'dataset_uid': a['dataset_uid'],
            'license': 1,
        })
        aid_counter += 1
        iid_counter += 1
    for neg_fno in duid2neg_frames_strict[a['dataset_uid']]:
        path = os.path.join(a['clip_uid'], "frame_{:07d}.png".format(neg_fno+1))
        assert os.path.isfile(os.path.join("/work/prannay/ego4d/images_val", path)), path
        coco_gt_data_images.append({
            'id': iid_counter,
            'visual_crop_id': [vcid_counter],
            'width': bb['original_width'],
            'height': bb['original_height'],
            'file_name': path,
            'dataset_uid': a['dataset_uid'],
            'license': 1,
        })
        iid_counter += 1
    vcid_counter += 1
coco_gt_data_categories = [{'id': 1, 'name': 'visual_crop'}]
coco_gt_data_info = {'description': 'ego4d'}
coco_gt_data_license = [{'url': 'http://creativecommons.org/licenses/by-nc-sa/2.0/', 'id': 1, 'name': 'Attribution-NonCommercial-ShareAlike License'}]
coco_gt_data = {
    'visual_crops': coco_gt_data_visual_crops,
    'annotations': coco_gt_data_annotations,
    'images': coco_gt_data_images,
    'categories': coco_gt_data_categories,
    'info': coco_gt_data_info,
    'licenses': coco_gt_data_license,
}
```
